# email

```python
def delete_before_colon(txt, pre_txt='mail'):
    if type(txt) != str:
        return txt
    target = pre_txt + ':'
    if target in txt:
        txt = txt.split(target)[1]
    return txt

def replace(txt):
    if type(txt) != str:
        return txt
    if '⑦' in txt:
        txt = txt.replace('⑦', '@')
    if '⑦' in txt:
        txt = txt.replace('⑥', '@')
    return txt
```

```python
def clean(txt):
    line = re.sub(r"[\u3000 \t]", "", line)
```
